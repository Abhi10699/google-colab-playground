```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("purchase_history.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
gender_encoded = pd.get_dummies(df['Gender'])
```

```python
gender_encoded
```

```python
gender_encoded = pd.get_dummies(df['Gender'], drop_first=True)
```

```python
gender_encoded
```

```python
df = pd.concat([df, gender_encoded], axis=1)
df
```

```python
x = df[['Male','Age','Salary','Price']].to_numpy()
x
```

```python
y = df['Purchased'].to_numpy()
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

```python
x_train[:10]
```

```python
from sklearn.neighbors import KNeighborsClassifier

K = 5
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(x_train, y_train)
```

```python
y_pred = knn.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
import pickle

# save scaler

with open("knn_model.pickle", 'wb') as f:
  pickle.dump(knn, f)
  f.close()

with open("knn_scaler.pickle", 'wb') as f:
  pickle.dump(scaler, f)
  f.close()
```
