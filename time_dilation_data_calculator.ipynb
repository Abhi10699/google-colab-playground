### Time Dilation Without the Effect of Gravity

```python
"""
escape_velocity - times speed of light
delta_t - seconds

The function returns time lag on the observers end

"""
def calculate_time_dilation(delta_t, escape_velocity):
    time_dilation = delta_t / math.sqrt(1 - (escape_velocity)**2)
    time_lag = delta_t - time_dilation
    return time_dilation, time_lag
```

```python
calculate_time_dilation(1, 0.9) # seconds
```

```python
sol = []
td_diff = []
td_obs = []

for i in range(1,10):
    nines = "9"*i
    final = float(f"0.{nines}")
    sol.append(final)

    # calculate time lag
    calcs = calculate_time_dilation(1, final)
    td_diff.append(abs(calcs[1]))
    td_obs.append(abs(calcs[0]))
```

```python
td_diff
```

```python
def plot_time_dilation(sol, td_obs):
    plt.plot(sol,td_obs)
    plt.plot(sol,[1]*len(sol))

#     plt.yscale("log")
#     plt.xscale("log")
    plt.title("Effect of time dillation as you approach the speed of light")
    plt.xlabel("Speed of light")
    plt.ylabel("Time lag felt by observer (in Seconds)")
    plt.show()
```

```python
plot_time_dilation(sol, td_obs)
```

### Time Dilation With The Effect Of Gravity

```python
G = 6.67430e-11 # gravitational constant
c = 299792458 # speed of light (m/s)

def time_dilation_due_to_gravity(M, r):
    dilation_factor = math.sqrt(1 - (2 * G * M) / (r * c**2))
    return dilation_factor
```

```python
import numpy as np
import matplotlib.pyplot as plt

def schwarzschild_radius(mass):
    # Schwarzschild radius formula: r = 2 * G * M / c^2
    G = 6.67430e-11  # Gravitational constant in m^3 kg^1 s^2
    c = 299792458  # Speed of light in m/s
    return 2 * G * mass / (c ** 2)

def energy_required(mass, M):
    G = 6.67430e-11  # Gravitational constant in m^3 kg^1 s^2
    r = schwarzschild_radius(M)
    return (G * mass * M) / r

# Mass of the black hole (in kg)
black_hole_mass = 1e25  # Example mass for the black hole (you can change this)

# Vary the mass of the object being compressed (from 1 kg to 1e25 kg)
object_masses = np.logspace(0, 25, 100, base=10)

# Calculate the energy required for each object mass
energies = [energy_required(mass, black_hole_mass) for mass in object_masses]

# Plot the chart
plt.figure(figsize=(10, 6))
plt.semilogx(object_masses, energies, label=f"Black Hole Mass: {black_hole_mass:.2e} kg")
plt.xlabel("Mass of the Object (kg)")
plt.ylabel("Energy Required (Joules)")
plt.title("Energy Required to Achieve Schwarzschild Radius")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def time_dilation_factor(M, R):
    G = 6.67430e-11
    c = 299792458.0
    return np.sqrt(1 - 2 * G * M / (c**2 * R))

def plot_time_dilation():
    # Mass of the black hole in kilograms
    M = 1e30  # Example mass, feel free to adjust this value

    # Range of distances from the center of the black hole in meters
    R = np.linspace(1000, 10**7, 1000)

    # Calculate time dilation factors for the range of distances
    time_dilation_factors = time_dilation_factor(M, R)

    # Plot the results
    plt.figure(figsize=(10, 6))
    plt.plot(R, time_dilation_factors)
    plt.xlabel('Distance from Black Hole (meters)')
    plt.ylabel('Time Dilation Factor')
    plt.title('Effect of Gravity on Time Dilation near a Black Hole')
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    plot_time_dilation()
```
